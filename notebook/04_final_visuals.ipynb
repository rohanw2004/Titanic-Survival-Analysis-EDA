# Titanic Dataset Final Analysis

## Objective

This notebook presents the most important findings discovered during exploratory analysis and feature engineering.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

```

```python
df = pd.read_csv(
    "../data/processed/feature_engineered_titanic.csv"
)

df.head()
```

# Gender vs Survival

```python
sns.countplot(
    x='Sex',
    hue='Survived',
    data=df
)

plt.title("Survival Based on Gender")

plt.show()
```

## Key Insight

Female passengers showed considerably higher survival rates compared to male passengers.

This suggests gender played an important role during evacuation.

# Passenger class vs Survival class

```python
sns.countplot(
    x='Pclass',
    hue='Survived',
    data=df
)

plt.title("Passenger Class vs Survival")

plt.show()
```

## Key Insight

First-class passengers had noticeably better survival rates compared to third-class passengers.

Passenger class appears to have influenced survival probability.

# Fare vs Survival

```python
sns.boxplot(
    x='Survived',
    y='Fare',
    data=df
)

plt.title("Fare Distribution by Survival")

plt.show()
```

## Key Insight

Passengers who survived generally paid higher fares compared to non-survivors.

This suggests economic factors may have influenced survival chances.

# Family Size vs Survival

```python
sns.countplot(
    x='FamilySize',
    hue='Survived',
    data=df
)

plt.title("Family Size vs Survival")

plt.show()
```

## Key Insight

Passengers traveling with smaller families generally showed better survival rates.

Very large families showed lower survival patterns.

# Correlation HeatMap

```python
numeric_df = df.select_dtypes(include='number')

plt.figure(figsize=(10,6))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")

plt.show()
```

## Key Insight

The heatmap shows relationships between numerical variables.

Passenger class and fare appear to show stronger relationships with survival compared to several other variables.

# Final Conclusion

From the analysis of the Titanic dataset:

- Female passengers showed significantly higher survival rates.
- Passenger class strongly influenced survival outcomes.
- Higher fare passengers generally showed better survival probability.
- Family size affected survival patterns, with smaller families performing better.
- Multiple factors together appear to influence survival rather than a single variable.

Overall, demographic, economic, and social factors appear associated with survival patterns observed in the dataset.
